Convert Marital_Status to String Labels

```python
data.rename(columns={'Orders': 'Order Count'}, inplace=True)
# Renaming the "Orders" column to "Order Count" for better readability.
```

```python
data['Gender'] = data['Gender'].replace({'F': 'Female', 'M': 'Male'})
# Replacing 'F' with 'Female' and 'M' with 'Male' for better readability in the Gender column
```

```python
# Mapping Marital_Status values to meaningful labels: 0 → 'Unmarried', 1 → 'Married'
data['Marital_Status'] = data['Marital_Status'].map({0: 'Unmarried', 1: 'Married'})
```

```python
print(data['Marital_Status'].unique()) #Verify the Change
```

```python
# Set figure size
plt.figure(figsize=(6,4))

# Create count plot with hue to properly apply the palette
sns.countplot(x='Gender', data=data, hue='Gender', palette=['blue', 'red'], legend=False)
plt.ylabel("order count")
# Add labels and title
plt.title('Gender Distribution of Customers', fontsize=14)
```

```python
# Creating a count plot for Gender with blue and red colors
ax = sns.countplot(x='Gender',data=data,palette=['blue', 'red'],hue='Gender')
plt.title("Gender-wise Distribution of Orders")
plt.ylabel("order count")
# Adding value labels on top of the bars for better readability
for bars in ax.containers:
    ax.bar_label(bars)
```

# Conclusion:
The analysis shows that female customers (7,828 orders) made significantly more purchases than male customers (3,403 orders) during the Diwali sales.
This indicates that women were the primary buyers, possibly due to a higher interest in festive shopping. 🚀

```python
data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#Aggregates the total purchase amount for each gender and sorts the results in descending order based on the amount spent.
```

```python
# Grouping data by Gender and summing the Amount
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Creating the bar plot
plt.figure(figsize=(6,4))
ax = sns.barplot(x='Gender', y='Amount', data=sales_gen, palette=['red', 'yellow'],hue='Gender')

# Adding value labels on the bars
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Total Sales Amount by Gender")
plt.show()
```

# Conclustion
From above graph we can see hat most of the buyers are female and even the purchasing power of females are greater than men

```python
# Check the actual column name
print(data.columns)  # Ensure 'Orders' exists

# If the column name was changed to 'Order Count', update accordingly
sales_state = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Adjust figure size for better visualization
sns.set(rc={'figure.figsize': (15, 5)})

# Create bar plot
ax=sns.barplot(data=sales_state, x='State', y='Amount')

# Show plot
plt.xticks(rotation=90)  # Rotate state names for better visibility
plt.title("Total Amounts by State")

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

These states likely benefit from a strong consumer base and higher purchasing power. Delhi and Madhya Pradesh show moderate sales, while states like Telangana, Punjab, and Rajasthan record significantly lower totals. This variation suggests regional differences in demand and market penetration. Businesses can focus on high-performing states for growth opportunities while implementing targeted strategies in lower-performing regions to boost sales.

# Product Category

```python
sns.set(rc={'figure.figsize': (20,5)})
ax = sns.countplot(data=data, x="Product_Category", palette="coolwarm",hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Distribution of Purchases Across Product Categories", fontsize=14)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# Conclusion:
The most purchased product categories are Clothing & Apparel, Food, and Electronics & Gadgets, indicating high consumer demand in these areas. Footwear & Shoes also has a notable number of purchases. On the other hand, categories like Auto, Stationery, Tupperware, and Veterinary products have significantly lower purchases, showing less consumer interest in these items.

```python
# Grouping data by Product Category and summing the Amount
data3 = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a bar plot with different colors
ax = sns.barplot(data=data3, x='Product_Category', y='Amount', palette="viridis",hue='Product_Category')  # 'viridis' gives a gradient color

# Adding labels to each bar
for bars in ax.containers:
    ax.bar_label(bars, fmt="%.2f")  # Formatting labels with 2 decimal places

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Adding title and labels
plt.title("Total Sales Amount by Product Category", fontsize=14)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Total Amount", fontsize=12)

# Displaying the plot
plt.show()
```

# Conclusion
The analysis shows that Food, Clothing & Apparel, and Electronics & Gadgets generate the highest sales, indicating strong customer demand in these categories. In contrast, Books, Decor, Pet Care, and Office Supplies have the lowest sales, suggesting lower consumer interest or purchase frequency. This insight can help businesses focus on marketing and inventory strategies. High-performing categories should be reinforced with better stock management and promotions, while low-performing ones may need targeted marketing. Understanding these trends allows for better decision-making in sales and product planning.

```python
data4 = data.groupby(['Product_ID'], as_index=False)['Order Count'].sum().sort_values(by='Order Count', ascending=False).head(10)

# Setting figure size for better visualization
sns.set(rc={'figure.figsize': (20, 5)})

# Creating a bar plot with different colors
ax = sns.barplot(data=data4, x='Product_ID', y='Order Count', palette="viridis",hue='Product_ID')

# Adding labels to bars
for bars in ax.containers:
    ax.bar_label(bars)

# Adding title
plt.title("Top 10 Products by Order Count")

# Displaying the plot
plt.show()
```

# Conclusion:
The bar chart displays the top 10 most ordered products based on order count. The product P00265242 is the most frequently purchased, with 127 orders, followed by P00110942 with 116 orders. The remaining products have order counts ranging from 91 to 74, indicating a gradual decline in popularity. This insight helps identify the best-selling products, which could be useful for inventory management, targeted marketing strategies, and promotional planning to boost sales further.

```python
import matplotlib.pyplot as plt

# Creating the figure and axis
fig, ax = plt.subplots(figsize=(12, 7))

# Grouping data by Product_ID, summing the Order Count, selecting the top 10, and sorting in descending order
top_products = data.groupby('Product_ID')['Order Count'].sum().nlargest(10).sort_values(ascending=False)

# Plotting the bar chart with custom colors
colors = plt.cm.viridis_r(range(0, 256, 25))  # Generating a gradient color palette
bars = top_products.plot(kind='bar', ax=ax, color=colors)

# Adding value labels on top of each bar
for bar in bars.patches:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,  # Adjust position
            int(bar.get_height()), ha='center', fontsize=12, fontweight='bold')

# Adding title and labels
ax.set_title("Top 10 Products by Order Count", fontsize=14, fontweight='bold')
ax.set_xlabel("Product ID", fontsize=12)
ax.set_ylabel("Order Count", fontsize=12)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

# Conclusion:
The bar chart showcases the Top 10 Products by Order Count, highlighting the most frequently ordered products. Product ID P00265242 stands out with the highest number of orders (127 orders), followed closely by P00110942 (116 orders). The top three products exhibit a significantly higher order volume compared to the rest, indicating their popularity among customers. The distribution suggests that a few products contribute to a large portion of total sales, which may be useful for inventory management and marketing strategies. Businesses can focus on these high-performing products to optimize stock and boost revenue.
